````python
# Month Count

month_list = []
for row in range (1, len(recweek_dict)+1,1):
    month_list.append(recweek_dict[str(row)]["Month"])
    
month_tally = {}

month_set = set(month_list)
for month in month_set:
    month_tally[month] = 0
    
for month in month_list:
    month_tally[month] += 1
    
# print(month_tally)

# Arranging the months and corresponding values

# used a number map so as to arrange 1-12 with the given word equivalent, and the month equivalent.
months = dict(one = "January", two = "February", three = "March", four = "April", five = "May", six = "June", seven = "July", eight = "August", nine = "September", ten = "October", eleven = "November", twelve = "December")
num_map = { "one" : 1, "two" : 2, "three": 3, "four": 4, "five": 5, "six": 6, "seven": 7, "eight": 8, "nine": 9,"ten": 10, "eleven": 11, "twelve": 12}

final_month_tally = {}
sorted(months, key = num_map.__getitem__)
for num in sorted(months, key = num_map.__getitem__):
    final_month_tally[months[num]] = month_tally[months[num]]
print(final_month_tally)
````

````python
# Day Count

day_list = []
for row in range (1, len(recweek_dict)+1,1):
    day_list.append(recweek_dict[str(row)]["Day"])
    
day_tally = {}

day_set = set(day_list)
for day in day_set:
    day_tally[day] = 0
    
for day in day_list:
    day_tally[day] += 1

# print(day_tally)

# Arranging the days and corresponding values

# Simply used a loop knowing that the values are successive from 1 - 28, with these values as keys. 
final_day_tally = {}
for day in range (1, len(day_tally)+1,1):
    final_day_tally[str(day)] = day_tally[str(day)]

print(final_day_tally)
````

````python
# Year Count

year_list = []
for row in range (1, len(recweek_dict)+1,1):
    year_list.append(recweek_dict[str(row)]["Year"])
    
year_tally = {}

year_set = set(year_list)
for year in year_set:
    year_tally[year] = 0
    
for year in year_list:
    year_tally[year] += 1

#print(year_tally)

# Arranging the years and corresponding values

# used a number map so as to arrange 1-5 with the given word equivalent, and the year equivalent.
years = dict(one= "1997", two= "1998", three= "1999", four= "2000", five= "2001") 
num_map = { "one" : 1, "two" : 2, "three" : 3, "four" : 4, "five" : 5}

final_year_tally = {}
sorted(years, key = num_map.__getitem__)
for num in sorted(years, key = num_map.__getitem__):
    final_year_tally[years[num]] = year_tally[years[num]]
print(final_year_tally)
````

````python
# Course Count

course_list = []
for row in range(1, len(recweek_dict)+1, 1):
    course_list.append(recweek_dict[str(row)]["Course"])
    
course_tally = {}

course_set = set(course_list)
for course in course_set:
    course_tally[course] = 0
    
for course in course_list:
    course_tally[course] += 1
    
print(course_tally)
````
