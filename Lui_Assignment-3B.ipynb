````python
# Calling of initial data frame and formulation of a copy for editing. 

import pandas as pd

dirty_df = pd.read_csv("recweekdata.csv")
#dirty_df

student_profile = pd.read_csv("studentprofiles.csv")
student_profile = student_profile.drop(columns = "lastname") # dropped lastname so as to avoid complications
#student_profile

cleaned_df = dirty_df.copy()
cleaned_df
````

````python
# Merging of Data Frames to the Copy for Manipulation

cleaned_df = pd.merge(dirty_df, student_profile, on = "firstname")
cleaned_df
````

````python
# Fixing the Month Format

month_dict = {"1" : "January",
              "2" : "February",
              "3" : "March",
              "4" : "April",
              "5" : "May",
              "6" : "June",
              "7" : "July",
              "8" : "August",
              "9" : "September",
              "10" : "October",
              "11" : "November",
              "12" : "December",
              "Jan" : "January",
              "Feb" : "February",
              "Mar" : "March",
              "Apr" : "April",
              "Jun" : "June",
              "Jul" : "July",
              "Aug" : "August",
              "Sep" : "September",
              "Oct" : "October",
              "Nov" : "November",
              "Dec" : "December",
              "January" : "January",
              "February" : "February",
              "March" : "March",
              "April" : "April",
              "May" : "May",
              "June" : "June",
              "July" : "July",
              "August" : "August",
              "September" : "September",
              "October" : "October",
              "November" : "November",
              "December" : "December"}

cleaned_df['birthmonth'] = cleaned_df.loc[:, "birthmonth"].apply(lambda x: month_dict[x])

cleaned_df
````

````python
# Fixing the Year Format

year_dict = {97:"1997",
             1997 : "1997",
             98:"1998",
             1998 : "1998",
             99:"1999",
             1999 : "1999",
             2000 : "2000",
             2001 : "2001"}

cleaned_df['birthyear'] = cleaned_df.loc[:, "birthyear"].apply(lambda x: year_dict[x])

cleaned_df
````

````python
# Fixing the Course Format

course_dict = {"BS LM" : "BS Legal Management",
               "LM" : "BS Legal Management",
               "Legal Management" : "BS Legal Management",
               "BS Legal Management":"BS Legal Management",
               "BS MGT" : "BS Management",
               "MGT" : "BS Management",
               "Management" : "BS Management",
               "BS Management":"BS Management",
               "BS ME" : "BS Management Engineering",
               "Management Engineering" : "BS Management Engineering",
               "ME" : "BS Management Engineering",
               "BS Management Engineering":"BS Management Engineering",
               "BS CTM" : "BS Communications Technology",
               "Communications Technology" : "BS Communications Technology",
               "CTM" : "BS Communications Technology",
               "BS Communications Technology":"BS Communications Technology",
               "BS ITE" : "BS IT Entrepreneurship",
               "ITE" : "BS IT Entrepreneurship",
               "Information Technology Entrepreneurship" : "BS IT Entrepreneurship",
               "BS IT Entrepreneurship":"BS IT Entrepreneurship"}

cleaned_df["course"] = cleaned_df.loc[:, "course"].apply(lambda x : course_dict[x])
cleaned_df
````

````python
# Fixing the Phone Number Format

cleaned_df.loc[:,"phonenumber"].replace(r' ', '',regex = True, inplace = True) 
cleaned_df.loc[:,"phonenumber"].replace(r'^9','+639',regex = True, inplace = True)

cleaned_df
````

````python
cleaned_df = cleaned_df.rename(index=str, columns={"firstname": "First_Name", "lastname": "Last_Name", "birthmonth" : "Birth_Month", "birthday": "Birth_Day", "birthyear": "Birth_Year", "course": "Course", "phonenumber": "Phone_Number", "gender": "Gender"})

cleaned_df.to_csv("cleanrecweekdata3b.csv", index = False)
````

````python
# To check if the unique values are correct, and simply left to arrange and collate. 

print(cleaned_df.Birth_Month.unique())
print(cleaned_df.Birth_Day.unique())
print(cleaned_df.Birth_Year.unique())
print(cleaned_df.Course.unique())
print(cleaned_df.Phone_Number.unique())
print(cleaned_df.Gender.unique())
````

````python
# Count for Number of people in each Birth Month

months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]

cleaned_df['Birth_Month'] = pd.Categorical(cleaned_df['Birth_Month'], categories=months, ordered=True)
cleaned_df = cleaned_df.sort_values(by='Birth_Month',ascending=True) # Arranges the Data Frame via birth month.

month_df = cleaned_df.groupby(["Birth_Month"])["First_Name"].count()

month_df

#cleaned_df
````

````python
# Count for Number of people in each Birth Day
# No longer needs sorting as done so by Pandas

day_df = cleaned_df.groupby(["Birth_Day"])["First_Name"].count()
day_df
````

````python
# Count for Number of people in each Birth Year
# No longer needs sorting as done so by Pandas

year_df = cleaned_df.groupby(["Birth_Year"])["First_Name"].count()
year_df
````

````python
# Count for Number of people in each Course

course_df = cleaned_df.groupby(["Course"])["First_Name"].count()
course_df
````

````python
# Count for Number of people subscribed to each Network Provider

provider = {"+63917":"Globe","+63918":"Smart", "+63927":"Globe"}

cleaned_df["Network_Provider"] = cleaned_df.loc[:,"Phone_Number"].str[0:6].apply(lambda x: provider[x])

provider_df = cleaned_df.groupby(["Network_Provider"])["First_Name"].count()
provider_df
````

````python
# Count for number of Male and Female Subscribers to Globe and Smart Respectively. 

provider_gender_df = cleaned_df.groupby(["Network_Provider", "Gender"])["First_Name"].count()

provider_gender_df
````

````python
stats_html = """

<html>
    <body style = "background-color = #66CCFF">
         <h1 style = "font-family:'Courier New'">Summary Statistics</h1>
         <div style = "color : #008000 ; border : 5px solid #008000">
             <h3 style = "font-size: 20px">People born in each month</h3>
                 <table style = "font-style:'Verdana';font-size: 15px">
                     <tr><th>Month</th><th>Count</th></tr>
                     {month}
                 </table>
         </div>
         <div style = "color : #008080 ; border : 5px solid #008080">
             <h3 style = "font-size: 20px">People born on each day</h3>
                 <table style = "font-style:'Verdana';font-size: 15px">
                     <tr><th>Day</th><th>Count</th></tr>
                     {day}
                 </table>
         </div>
         <div style = "color : #009900 ; border : 5px solid #009900">
             <h3 style = "font-size: 20px">People born on each year</h3>
                 <table style = "font-style:'Verdana';font-size: 15px">
                     <tr><th>Year</th><th>Count</th></tr>
                     {year}
                 </table>
         </div>
         <div style = "color : #0000FF ; border : 5px solid #0000FF">
             <h3 style = "font-size: 20px">People coming from each course</h3>
                 <table style = "font-style:'Verdana'; font-size: 15px">
                     <tr><th>Course</th><th>Count</th></tr>
                     {course}
                 </table>
         </div>
         <div style = "color : #000080 ; border : 5px solid #000080">
            <h3 style = "font-size: 20px">People having numbers from Globe and Smart</h3>
                 <table style = "font-style:'Verdana'; font-size: 15px">
                     <tr><th>Provider</th><th>Count</th></tr>
                     {provider}
                 </table>
         </div>
         <div style = "color : #003300 ; border : 5px solid #003300">
             <h3 style = "font-size: 20px">Provider - Gender Ratio</h3>
                 <table style = "font-style:'Verdana'; font-size: 15px">
                     <tr><th>Provider</th><th>Gender</th><th>Count</th></tr>
                     {p_g_ratio}
                 </table>
         </div>
    </body>
</html>
"""

# Note that Series can be looped through right away. 
months_table = ""
for month in month_df.keys():
    months_table += "<tr><td>{}</td><td>{}</td></tr>".format(month, month_df[month])

day_table = ""
for day in day_df.keys():
    day_table += "<tr><td>{}</td><td>{}</td></tr>".format(day, day_df[day])
    
year_table = ""
for year in year_df.keys():
    year_table += "<tr><td>{}</td><td>{}</td></tr>".format(year, year_df[year])
    
course_table = ""
for course in course_df.keys():
    course_table += "<tr><td>{}</td><td>{}</td></tr>".format(course, course_df[course])

provider_table = ""
for provider in provider_df.keys():
    provider_table += "<tr><td>{}</td><td>{}</td></tr>".format(provider, provider_df[provider])
    
# Note the Multi Index Levels of the double grouped content. 
provider_gender_table = ""
for content in provider_gender_df.keys():
    provider_gender_table += "<tr><td>{}</td><td>{}</td><td>{}</td></tr>".format(content[0], content[1],provider_gender_df[content])
    

with open('stats3b.html',"w+") as file:
    content_html = stats_html.format(month = months_table, day = day_table, year = year_table, course = course_table, provider = provider_table, p_g_ratio = provider_gender_table)
    file.write(content_html)
````
